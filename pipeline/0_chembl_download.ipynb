```python
from chembl_webresource_client.new_client import new_client

# Define the organism(s) you're interested in (e.g., Mycobacterium tuberculosis)
organism_name = 'Mycobacterium tuberculosis'

# Initialize a target resou1rce
target_resource = new_client.target

# Use the `search` method to search for targets associated with the organism
targets = target_resource.search(organism_name)

# Filter targets to include only those explicitly containing "Mycobacterium tuberculosis"
filtered_targets = [target for target in targets if "Mycobacterium tuberculosis".lower() in target['organism'].lower()]

id_name = {}
# Print the filtered targets
for i, target in enumerate(filtered_targets):
    print(i+1, '\t\t', target['target_chembl_id'], target['pref_name'])
    id_name[target['target_chembl_id']] = target['pref_name']

#save id_name to a file
import yaml
with open('../data/chemblapi/full/id_name.yaml', 'w') as f:
    yaml.dump(id_name, f)
```

```python
from chembl_webresource_client.new_client import new_client


import yaml
with open('../data/chemblapi/mics/id_name.yaml', 'w') as f:
    yaml.dump(id_name, f)

activity = new_client.activity

for i, target in enumerate(filtered_targets):
    print('Processing', target['pref_name'], '...', i+1, 'of', len(filtered_targets))
    chembl_id = target['target_chembl_id']
    bioactivities = activity.filter(target_chembl_id=chembl_id).filter(standard_type="MIC")
    #load to pandas dataframe to save as csv
    df = pd.DataFrame.from_dict(bioactivities)
    df.to_csv(f'../data/chemblapi/mics/{chembl_id}.csv')
```
